```python
URL = "https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD"
```

```python
from urllib.request import urlretrieve
```

```python
urlretrieve(URL,"Fremont.csv")
```

```python
with open("Fremont.csv") as f:
    for _ in range(5): # first 5 lines
        print(f.readline())
```

```python
import pandas as pd
data = pd.read_csv('Fremont.csv',index_col = "Date", parse_dates=['Date'])
data.head()
```

```python
data.plot()
```

```python
data.resample('W').sum().plot()
```

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn")
data.columns = ["Total","East","West"]
data.resample('W').sum().plot()
```

```python
ax = data.resample('D').sum().rolling(365).sum().plot()
ax.set_ylim(0,None)
```

```python
data.groupby(data.index.time).mean().plot();
```

```python
pivoted = data.pivot_table('Total', index =data.index.time, columns = data.index.date)
pivoted.iloc[:5,:5]
```

```python
pivoted.plot(legend=False, alpha = 0.01)
```

Hypothesis :  commute days have bike counts peaks around the morning and at end of work hours, while *weekends* and *holidays* have a smoother peak
